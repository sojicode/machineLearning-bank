### 3. Use z-score normalization 

```python
# Use z-score to std values of the income attribute
# Do not chagne the original age attribute or add discretized age to the table.

bk_income = bk_table["income"] 
mean_income = bk_table["income"].mean()
std_income = bk_table["income"].std()

zscore_income = (bk_income - mean_income) / std_income
zscore_income.head(10)
```

```python
# other way

z_score = lambda x: (x - x.mean()) / x.std() if (x.dtypes==np.float64 or x.dtypes==np.int64) else x
bk_std = bk_table.copy()
zscore_bk = bk_std.apply(z_score)
zscore_income = zscore_bk["income"]
zscore_income.head(10)
```

### 5. Min-Max Normalization to transform the values of all numeric attributes 

```python
bk_numeric = bk_table[["income","age","children"]]
bk_numeric.head()
```

```python
min_num = bk_table[["income","age","children"]].min()
max_num = bk_table[["income","age","children"]].max()
print(min_num, max_num)
```

```python
min_max_norm = (bk_table[["income","age","children"]] - min_num) / (max_num - min_num)
min_max_norm.head(10)
```

```python
# transform the values of (incom,age, children) in the original table

bk_table[["income","age","children"]] = min_max_norm[["income","age","children"]]
bk_table.head()
```
